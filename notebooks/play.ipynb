```python
import torch, torch.nn as nn
import matplotlib.pyplot as plt

# tiny conv net
net = nn.Sequential(
    nn.Conv2d(1, 8, 3, padding=1), nn.ReLU(), # 51x51x1 -> 51x51x8 | receptive field: 3x3
    nn.Conv2d(8, 8, 3, padding=1), nn.ReLU(), # 51x51x8 -> 51x51x8 | receptive field: 5x5
    nn.Conv2d(8, 1, 3, padding=1) # 51x51x8 -> 51x51x1 | receptive field: 7x7
)

x = torch.zeros(1,1,51,51, requires_grad=True)
out = net(x)
center = out[0,0,25,25]
center.backward()

plt.imshow(x.grad[0,0].abs(), cmap='hot')
plt.title("Effective receptive field heatmap")
plt.show()
```
